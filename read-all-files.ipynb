```python
import glob
import pandas as pd
```

```python
# you can use glob with slicing
for file in glob.glob("mix20_rand700_tokens_0211/tokens/pos/*.txt")[:10]:
    print(file)
```

```python
reviews = []
for f in glob.glob("mix20_rand700_tokens_0211/tokens/pos/*.txt")[:10]:
    with open(f, 'r', encoding='latin1') as file:
        pos_review = file.read()
        reviews.append((pos_review, 1))
        
for f in glob.glob("mix20_rand700_tokens_0211/tokens/neg/*.txt")[:10]:
    with open(f, 'r', encoding='latin1') as file:
        pos_review = file.read()
        reviews.append((pos_review, 0))
```

```python
len(reviews)
```

```python
df_reviews = pd.DataFrame(reviews, columns=['text', 'review'])
```

```python
df_reviews
```

```python
# add columns
df_reviews['alpha_text'] = df_reviews.text.str.replace('[^a-zA-Z]', ' ')
```
