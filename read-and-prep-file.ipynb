```python
stop_words = ['a', 'the', 'by', 'and', 'on', 'for', 'is', 'to', 'in', 'of']
```

```python
# what if you don't want...
pos_words = []
for w in pos_review.split():
    w = re.sub(r'\W+', '', w)
    if len(w) > 0 and w not in stop_words:
        pos_words.append(w)
```

```python
# what about stemming and lemmatizing?
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# there will be a difference
```

```python
print(stemmer.stem("narrators"))
print(lemmatizer.lemmatize("narrators"))
```

```python
print(stemmer.stem("narration"))
print(lemmatizer.lemmatize("narration"))
```

```python
# what if you don't want...
pos_words = []
for w in pos_review.split():
    w = re.sub(r'\W+', '', w)
    if len(w) > 0 and w not in stop_words:
        pos_words.append(stemmer.stem(w))
```

```python
print(' '.join(pos_words))
```
